```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Function to plot a circle
def plot_circle(ax, center, radius, color, z=0):
    phi = np.linspace(0, 2*np.pi, 100)
    x = center[0] + radius * np.cos(phi)
    y = center[1] + radius * np.sin(phi)
    ax.plot(x, y, zs=z, zdir='z', color=color)

# Function to plot a smiley face
def plot_smiley_face():
    # Create a new figure and 3D axis
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot face outline
    plot_circle(ax, [0, 0], 1, 'yellow')
    
    # Plot eyes 
    plot_circle(ax, [-0.35, 0.35], 0.1, 'black')
    plot_circle(ax, [0.35, 0.35], 0.1, 'black')
    
    # Plot smile
    phi = np.linspace(0, np.pi, 100)
    x = 0.6 * np.cos(phi)
    y = -0.2 + 0.2 * np.sin(-phi)
    ax.plot(x, y, zs=0, zdir='z', color='black')
    
    # Show the plot
    plt.show()

# Call the function to plot the smiley face
plot_smiley_face()
```
